```python
real_lidar_path="/data/nuscenes/nuscenes-full/samples/LIDAR_TOP/n015-2018-10-08-15-36-50+0800__LIDAR_TOP__1538984246948247.pcd.bin"
pointcloud = np.fromfile((real_lidar_path), dtype=np.float32, count=-1).reshape([-1, 5])
```

```python
def points_vis(geom1):
    fig = mayavi.mlab.figure(bgcolor=(0, 0, 0), size=(1920, 1080))
    mayavi.mlab.points3d(geom1[:, 0], geom1[:, 1], geom1[:, 2],
                         geom1[:, 2],  # Values used for Color
                         mode="point",
                         colormap='spectral',  # 'bone', 'copper', 'gnuplot'
                         # color=(0, 1, 0),   # Used a fixed (r,g,b) instead
                         figure=fig,
                         )
    mayavi.mlab.show()
```

```python
points_vis(pointcloud)
```

```python
virtual_lidar_path="/data/nuscenes/virtual_points-yolov5+depthmap/2/LIDAR_TOP/n015-2018-10-08-15-36-50+0800__LIDAR_TOP__1538984246948247.pcd.bin.pkl.npy"
v_pointcloud=np.load(virtual_lidar_path)
points_vis(v_pointcloud)
```

```python
def points_vis_two(geom1, geom2):
    print(geom1.shape)
    print(geom2.shape)
    fig = mayavi.mlab.figure(bgcolor=(0, 0, 0), size=(1920, 1080))
    mayavi.mlab.points3d(geom1[:, 0], geom1[:, 1], geom1[:, 2],
                         geom1[:, 2],  # Values used for Color
                         mode="point",
                         colormap='spectral',  # 'bone', 'copper', 'gnuplot'
                         color=(0, 1, 0),  # Used a fixed (r,g,b) instead
                         figure=fig,
                         )
    mayavi.mlab.points3d(geom2[:, 0], geom2[:, 1], geom2[:, 2],
                         geom2[:, 2],  # Values used for Color
                         mode="point",
                         colormap='spectral',  # 'bone', 'copper', 'gnuplot'
                         #                          color=(1, 0, 0),   # Used a fixed (r,g,b) instead
                         figure=fig,
                         )

    mayavi.mlab.show()
```

```python
points_vis_two(pointcloud, v_pointcloud)
```
